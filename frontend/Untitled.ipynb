```python
import pandas as pd
import re
def get_l1cache(data, type):
    if type == 'D':
        match = re.match(r".+?\+\s+(\d+)", data)
        if match:
            return match.group(1)
    else:
        match = re.match(r"\s*(\d+)", data)
        if match:
            return match.group(1)
def get_l2cache(data):
    match = re.match(r"\s*(\d+\.?\d*)\s*(\S+)", data)
    if match:
        if match.group(2) == 'KB':
            return float(match.group(1)) / 1024
        else:
            return match.group(1)
def get_l3cache(data):
    match = re.match(r"\s*(\d+\.?\d*)\s*(\S+)", data)
    if match:
        return match.group(1)
    
pd.options.display.max_columns = None
pd.set_option('display.min_rows',1)
df = pd.read_csv('added_scores_details.csv', low_memory=False)
```

```python
cpu_df = df.iloc[:,[2, 10, 11, 8,7 , 9, 16, 17, 18, 14, 15, 12, 13,16, 17, 18]]

cpu_df.insert(3, "CoresPerCPU",0)
cpu_df.insert(8, "L1D cache(KB)", 0)
cpu_df.insert(9, "L1I cache(KB)", 0)
cpu_df.insert(11, "L2 cache(MB)", 0)
cpu_df.insert(13, "L3 cache(MB)", 0)

for i in range(len(cpu_df)):
    cpu_df.iloc[i,3] = int(cpu_df.iloc[i,5] / cpu_df.iloc[i,4])
    cpu_df.iloc[i,8] = get_l1cache(cpu_df.iloc[i,7],'D')
    cpu_df.iloc[i,9] = get_l1cache(cpu_df.iloc[i,7],'I')
    cpu_df.iloc[i,11] = get_l2cache(cpu_df.iloc[i,10])
    cpu_df.iloc[i,13] = get_l2cache(cpu_df.iloc[i,12])
cpu_df = cpu_df.drop(columns=[cpu_df.columns[4], cpu_df.columns[5]]).copy()#.drop_duplicates()
cpu_df.to_csv("CPU维表.csv",index=False,encoding='utf_8_sig')
cpu_df
```

```python
server_df = df.iloc[:, [0, 1, 2, 3, 4, 7, 8, 10, 11, 16, 20, 21, 22, 23, 24 ,9, 16, 17, 18] ]

#返回列表【内存大小，规模，粒度】
def get_mem(data):
    match = re.match(r"(\d+)\s*([A-Za-z]{2})\s*\(\s*(\d+\s*x\s*\d+)\s*([A-Za-z]{2})\s*(\d+.*x\d+)?",data)
    if match:
        if match.group(2) == 'GB':
            vol = float(match.group(1)) / 1024
        return [match.group(1), match.group(3),  match.group(5)]
#返回列表【硬盘大小，GB/TB，是否SSD】
def get_store(data):
    match = re.match(r".*?(\d+\.?\d*)\s*(T|G)B?", data)
    if match:
        return [match.group(1), match.group(2)+"B", bool(data.find('SSD'))]
# 返回列表【发行版本，版本号，内核版本】
def get_op(data):
    match = re.match(r".*([345]\.\d\d?\.\d\d?)", data)
    if match:
        kernel_v = match.group(1)
        match = re.match(r"^([A-Za-z ]+)\s+(\d+\.?\d*)", data)
        if match:
            rel = match.group(1)
            rel_v = match.group(2)
            rel.replace("release", "")
            rel.replace(",", "")
            return [rel, rel_v, kernel_v]
    else:
        return data


# 返回列表【C/C++，ver, Fortran, ver】
def get_compiler(data):
    match = re.match(r"C/C\+\+/Fortran", data)
    if match:
        match = re.match(r".*Version\s*(\d+\.\d+\.?\d?\d?\d?\.?\d?\d?\d?)\s*of\s*([A-Za-z0-9 ]+)", data)
        if match:
            return [match.group(2), match.group(1), match.group(2), match.group(1)]
    else:
        cidx = data.find(r"C/C++")
        fidx = data.find(r"Fortran")
        if cidx > fidx:
            fstr = data[:cidx-1]
            cstr = data[cidx:]
        else:
            cstr = data[:fidx-1]
            fstr = data[fidx:]

        cstr = cstr.replace(",","")
        cstr = cstr.replace(";","")
        fstr = fstr.replace(",","")
        fstr = fstr.replace(";","")
   
        match = re.match(r".*C/C\+\+:\s*Version\s*(\d+\.\d+\.?\d*\.?\d*)\s*(for|of)\s*(.+)", cstr)
       
        if match:
            cc = match.group(3)
            cc_v = match.group(1)

        match = re.match(r".*Fortran:\s*Version\s*(\d+\.\d+\.?\d*\.?\d*)\s*(for|of)\s*(.+)", fstr)      
        if match:
            fortran = match.group(3)
            fortran_v = match.group(1)
        return [cc, cc_v, fortran, fortran_v]
    
server_df.insert(10, "内存大小",0)
server_df.insert(11, "内存条",0)
server_df.insert(12, "颗粒度",0)

server_df.insert(14, "硬盘大小",0)
server_df.insert(15, "GB/TB",0)
server_df.insert(16, "是否固态",0)

server_df.insert(18, "发行版本", 0)
server_df.insert(19, "版本号", 0)
server_df.insert(20, "内核版本", 0)

server_df.insert(23, "C/C++编译器", 0)
server_df.insert(24, "C/C++编译器版本", 0)
server_df.insert(25, "Fortran编译器", 0)
server_df.insert(26, "Fortran编译器版本", 0)

for i in range(len(server_df)):
    #print(server_df.iloc[i,9])
    mem = get_mem(server_df.iloc[i,9])
    server_df.iloc[i,10] = mem[0]
    server_df.iloc[i,11] = mem[1]
    server_df.iloc[i,12] = mem[2]
    
    #print(server_df.iloc[i,13])
    stor = get_store(server_df.iloc[i,13])
    server_df.iloc[i,14] = stor[0]
    server_df.iloc[i,15] = stor[1]
    server_df.iloc[i,16] = stor[2]
    
    #print(server_df.iloc[i,17])
    op = get_op(server_df.iloc[i,17])
    if type(op) == list:
        server_df.iloc[i,18] = op[0]
        server_df.iloc[i,19] = op[1]
        server_df.iloc[i,20] = op[2]
    else:
        server_df.iloc[i,18] = server_df.iloc[i,17]
        server_df.iloc[i,19] = ""
        server_df.iloc[i,20] = ""
        
    #print(server_df.iloc[i,22])
    compiler = get_compiler(server_df.iloc[i,22])
    server_df.iloc[i,23] = compiler[0]
    server_df.iloc[i,24] = compiler[1]
    server_df.iloc[i,25] = compiler[2]
    server_df.iloc[i,26] = compiler[3]

server_df = server_df#.drop_duplicates()
server_df.to_csv("服务器维表.csv",index=False,encoding='utf_8_sig')
server_df
```
